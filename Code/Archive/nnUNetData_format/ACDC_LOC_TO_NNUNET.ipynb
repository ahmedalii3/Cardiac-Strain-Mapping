# Make the masks inner and outer contours

```python
masks_path = "/Users/ahmed_ali/Documents/GitHub/GP-2025-Strain/Data/ACDC_Loc/Masks"
Masks = os.listdir(masks_path)
```

```python
for mask in Masks:
    mask_file = mask
    mask = np.load(os.path.join(masks_path, mask))
    mask[mask == 1] = 0
    mask[mask == 2] = 1
    mask[mask == 3] = 2 
    # save the mask
    np.save(os.path.join(masks_path, mask_file), mask)
    print(mask_file, "is done")
```

# convert the np to nii.gz files


```python
Images_path = "/Users/ahmed_ali/Documents/GitHub/GP-2025-Strain/Data/ACDC_Loc/Images"
Images = os.listdir(Images_path)
masks_path = "/Users/ahmed_ali/Documents/GitHub/GP-2025-Strain/Data/ACDC_Loc/Masks"
Masks = os.listdir(masks_path)
```

```python
images_nii = "/Users/ahmed_ali/Documents/GitHub/GP-2025-Strain/Data/ACDC_Loc/Images_nii"
masks_nii = "/Users/ahmed_ali/Documents/GitHub/GP-2025-Strain/Data/ACDC_Loc/Masks_nii"

for i, img in enumerate(Images):
    image_file = img

    image = np.load(os.path.join(Images_path, img))
    mask = np.load(os.path.join(masks_path, img))

    affine = np.eye(4)

    nii_img = nib.Nifti1Image(image, affine)
    nii_mask = nib.Nifti1Image(mask, affine)

    nib.save(nii_img, os.path.join(images_nii, f"case_{i+1}_0000.nii.gz"))
    nib.save(nii_mask, os.path.join(masks_nii, f"case_{i+1}.nii.gz"))

    print(image_file, "is done")
    
```
