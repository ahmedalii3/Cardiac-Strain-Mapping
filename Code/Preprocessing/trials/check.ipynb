```python
#plot the numpy array mask and image

import numpy as np
import matplotlib.pyplot as plt

def plot_numpy_array(numpy_file):
    numpy_array = np.load(numpy_file)
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(numpy_array[0], cmap="gray")
    ax[0].axis("off")
    ax[0].set_title("Image")
    ax[1].imshow(numpy_array[1], cmap="gray")
    ax[1].axis("off")
    ax[1].set_title("Mask")
    plt.show()

# Example usage:
numpy_file = '/Users/osama/GP-2025-Strain/Data/ACDC/train_numpy/patient001/patient001_frame01_slice_1_ACDC.npy'
plot_numpy_array(numpy_file)
```
