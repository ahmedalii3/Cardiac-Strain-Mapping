```python
import os
import sys
import shutil
```

```python
simulated_data_frames_path = "/Users/ahmed_ali/Documents/GitHub/GP-2025-Strain/Data/Simulated_V2.1/Frames"
directories = os.listdir(simulated_data_frames_path)
print(directories)
```

```python
patients_dictionary = {}
for dir in directories:
    
    
    file = os.path.join(simulated_data_frames_path, dir)
    patient_number = dir.split("_")[0]
    if patient_number not in patients_dictionary:
        patients_dictionary[patient_number] = []
        # print(patient_number)
    patients_dictionary[patient_number].append(file)


print(patients_dictionary)
```

```python
# create folder for each patient
patients_path = "/Users/ahmed_ali/Documents/GitHub/GP-2025-Strain/Data/Simulated_V2.1/Frames_patients"
for patient in patients_dictionary:
    patient_directory = os.path.join(patients_path, patient)
    os.makedirs(patient_directory, exist_ok=True)
    for file in patients_dictionary[patient]:
        print(file)
        print(patient_directory)
        shutil.copy(file, patient_directory)
    print(f"Patient {patient} done")
```
