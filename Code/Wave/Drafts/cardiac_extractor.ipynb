```python
import numpy as np
import matplotlib.pyplot as plt

def plot_and_overlay_numpy_array(numpy_file):
    # Load the numpy file
    numpy_array = np.load(numpy_file)
    
    # Create subplots
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))
    
    # Plot the original image
    ax[0].imshow(numpy_array[0], cmap="gray")
    ax[0].axis("off")
    ax[0].set_title("Image")
    
    # Plot the mask
    ax[1].imshow(numpy_array[1], cmap="gray")
    ax[1].axis("off")
    ax[1].set_title("Mask")
    
    # Plot the overlay with reduced opacity for the first image
    ax[2].imshow(numpy_array[0], cmap="gray", alpha=0.5)
    ax[2].imshow(numpy_array[1], cmap="jet", alpha=0.5)  # Use "jet" colormap for contrast
    ax[2].axis("off")
    ax[2].set_title("Overlay (Image + Mask)")
    
    # Display the plot
    plt.show()

# Example usage:
numpy_file = '/Users/osama/GP-2025-Strain/Data/ACDC/train_numpy/patient001/patient001_frame01_slice_3_ACDC.npy'
plot_and_overlay_numpy_array(numpy_file)
```
