```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
strain_values = np.load("/Users/osama/GP-2025-Strain/Code/Wave/strain_values.npz")
```

```python
# Extract keys and corresponding values
keys = list(strain_values.keys())
values = [strain_values[key].flatten() for key in keys]

# Create the plot
plt.figure(figsize=(10, 6))
plt.boxplot(values, vert=True, patch_artist=True)

# Set x-axis labels
plt.xticks(range(1, len(keys) + 1), keys, rotation=45, ha="right")

# Labels and title
plt.xlabel("Keys")
plt.ylabel("Strain Values")
plt.title("Strain Values Distribution per frame")

plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
